```python
# linear regression formula for y = thetha0 + thetha1*x1 + thetha2 * x2 ....
```

```python
def hypothesis(X,theta):   #x was small , it should have been capital 
    y_=0
    n=X.shape[0]

    for i in range(n):
        y_+=theta[i]*X[i]

    return y_
    
```

```python
def cost(X,Y,theta):
    m=X.shape[0]
    total_error=0

    for i in range(m):
        y_p=hypothesis(X[i],theta)

        total_error+= (y_p - Y[i])**2

    return total_error/m
```

```python
def gradient(X,Y,theta):
    m,n = X.shape   #should bw x.shape as well

    grad= np.zeros((n,))

    for j in range(n):
        for i in range(m):
            x=X[i]
            y_p=hypothesis(X[i],theta)
            y_t=Y[i]

            grad[j]+= (y_p-y_t)*x[j]
    return grad/m
```

```python
def gradient_descent(X, Y , learning_rate=0.001, max_steps=2000):
    m,n = X.shape       #changed from shape[0] as the answer needs to be 2d as well to accomdate 2 variables rather than 1d
    theta= np.zeros((n,))
    cost_epoch=[]

    for i in range(max_steps):
        grad=gradient(X,Y, theta)
        e=cost(X,Y, theta)
        cost_epoch.append(e)

        for j in range(n):
            theta[j]=theta[j]-learning_rate*grad[j]

    return theta, cost_epoch
```
