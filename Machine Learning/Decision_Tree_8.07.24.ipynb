```python
# Entropy 
```

```python
def Entropy(col):
    favour= np.unique(col,return_counts=True)  # unique values nikal raha hai yeh
    N= col.shape   # total number of values 
    ent_py=0    # initally set to 0
    for i in favour[1]:    # for every unique value
        prob = i/N      # uski probability
        ent_py += -1 * (prob * np.log2(prob))
    return ent_py
    
```

```python
def divide_data(X_data, feat, value):
    X_left = X_data[X_data[feat] < value]    # selects all rows that are less than value from the given dataset
    X_right = X_data[X_data[feat]>= value]
    return X_left, X_right
```

```python
# Information gain 
```

```python
def info_gain(X_data, feat, value):
    left,right= divide_data(X_data, feat, value)  # divide into left and right

    prob_l= left.shape[0]/X_data.shape[0]   # getting probabilites for both sides, understanding the amount divided into each of the two 
    prob_r= right.shape[0]/X_data.shape[0]

    i_gain= Entropy(X_data.Survived) - (prob_l * Entropy(left.Survived) + prob_r * Entropy(right.Survived))  # finding info gain by subtracting the weighted sum of the entropies of left and right from the entropy of the data 

    return i_gain
    # helps us figure out which feature should be used for dividing the dataset   
```

```python
class Decision_Tree:
    def __init__(self, depth=0, max_depth=7):
        self.left= None
        self.right= None
        self.feat= None
        self.value = None
        self.max_depth= max_depth
        self.depth= depth
        self.target= None         # initializing a class

    def train(self,X_train):
        f = ['Pclass',	'Sex'	,'Age'	,'SibSp'	,'Parch',	'Fare'	,'Embarked']  # reintroducing the colums
        i_gain=[]   # information gain storing list, empty currently

        for i in f:
            i_gain.append(info_gain(X_train, i , X_train[i].mean()))   # adding all the infomation gains for each of the columns

        self.feat= f[np.argmax(i_gain)]   # choosing the feature with the maximum information gain 
        self.value= X_train[self.feat].mean()  # choosing its mean as the threshold as well 

        data_left, data_right= divide_data( X_train, self.feat, self.value)
        data_left= data_left.reset_index(drop= True)
        data_right= data_right.reset_index(drop= True)

        print("Making tree with feature:", self.feat, "Depth:", self.depth)

        #base cases, which work with the smallest units 
        if data_right.shape[0]==0 or data_left.shape[0]==0:
            if X_train.Survived.mean()>=0.5:           # for leaf nodes , where no more children are present 
                self.target= 'Survived'
            else:
                self.target='Not Survived'
            return 'Done Training'

        if self.depth>= self.max_depth:
            if X_train.Survived.mean()>=0.5:
                self.target= 'Survived'       # for whent the max depth is exceeded by the depth 
            else:
                self.target='Not Survived'
            return 'Done Training'

        self.left= Decision_Tree(depth= self.depth+1)
        self.left.train(data_left)                             # recursion for each of the nodes occurs here 

        self.right= Decision_Tree(depth= self.depth+1)
        self.right.train(data_right)

        if X_train.Survived.mean()>=0.5:
            self.target= 'Survived'
        else:
            self.target='Not Survived'
        return 'Done Training'

    def predict(self, test):
        if test[self.feat]>= self.value:
            if self.right is None:
                return self.target                            # predicting values, if higher than threshold it moves to the right sub tree and if not 
            return self.right.predict(test)                    # it moves to the left sub tree 
        else:
            if self.left is None:
                return self.target
            return self.left.predict(test)

```
