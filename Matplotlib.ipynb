```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
var1 = np.arange(0, 10)
```

```python
var2 = var1**2
```

```python
print(var2, var1)
```

```python
plt.plot(var1, color = 'red', label = 'Price of gold', linestyle = '--')
plt.plot(var2, color = 'blue', label = 'Price of oil', linestyle = '-.')
plt.legend(title = 'Difference in Price')
plt.xlabel('Time')
plt.ylabel('Prices in $')
plt.title('Time vs Prices')
```
